DUPLICATES()

```python
import pandas as pd

df = pd.read_csv('uce.csv')

# Iterate over each column and check for duplicates
for column in df.columns:
    duplicates = df[column].duplicated()
    num_duplicates = duplicates.sum()
    
    if num_duplicates > 0:
        print(f"Column '{column}' has {num_duplicates} duplicate(s).")
        print("Duplicate rows:")
        print(df[duplicates])
        print()
```

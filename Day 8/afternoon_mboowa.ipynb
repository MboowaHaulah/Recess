2.Write a Pandas program to add, subtract, multiply and divide two Pandas series. 
Sample Series: [2,4,6,8,14], [1,3,5,7,9]

```python

import pandas as pd

series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])

addition = series1 + series2
subtraction = series1 - series2
multiplication = series1 * series2
division = series1 / series2

print("Addition:")
print(addition)
print("Subtraction:")
print(subtraction)
print("Multiplication:")
print(multiplication)
print("Division:")
print(division)
```

```python
#5.Write a Pandas program to display the most frequent value in the given series and replace everything else as ‘Other’ in the series. 
#(Use the ‘HomeTeamGoals’ column in the dataset provided)
import pandas as pd

df = pd.read_csv('C:/Users/USER/Desktop/RECESS/Day/cup.csv')  

# Find the most frequent value
most_frequent_value = df['HomeTeamGoals'].mode()[0]

# Replace everything else as 'Other'
df['HomeTeamGoals'] = df['HomeTeamGoals'].apply(lambda x: most_frequent_value if x == most_frequent_value else 'Other')

print(df['HomeTeamGoals'])
```

```python
#12.Write a Pandas program to add 2 rows to your DataFrame.
import pandas as pd

# Create a DataFrame 'cup' with some initial data
df = pd.DataFrame({
    'HomeTeam': ['Team X', 'Team Y'],
    'AwayTeam': ['Team Z', 'Team W'],
    'HomeTeamGoals': [3, 1],
    'AwayTeamGoals': [2, 0]
})

# Define new rows to be added
new_rows = [{'HomeTeam': 'Team A', 'AwayTeam': 'Team B', 'HomeTeamGoals': 2, 'AwayTeamGoals': 1},
            {'HomeTeam': 'Team C', 'AwayTeam': 'Team D', 'HomeTeamGoals': 0, 'AwayTeamGoals': 2}]

# Append the new rows to the DataFrame
df = pd.concat([df, pd.DataFrame(new_rows)], ignore_index=True)

# Print the updated DataFrame
print(df)
```

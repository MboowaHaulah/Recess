```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

```

```python
#IGNORE WARNINGS
import warnings
warnings.filterwarnings("ignore")

# Load the heart dataset 
df = pd.read_csv("heart.csv")

y = df.target.values
x_data = df.drop(['target'], axis=1)

# Split the data 
X = df.drop("target", axis=1)
y = df["target"]
```

```python
# SPLIT THE DATA FOR TRAINING '67'AND TESTING'33'
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_data, y, test_size=0.33, random_state=42)
```

```python
# Initialize logistic regression model with a different solver
logistic_regression_model = LogisticRegression(solver='saga', max_iter=1000).fit(X_train, y_train)

# Initialize and train models
logistic_regression_model = LogisticRegression().fit(X_train, y_train)
knn_model = KNeighborsClassifier().fit(X_train, y_train)
svm_model = SVC().fit(X_train, y_train)
naive_bayes_model = GaussianNB().fit(X_train, y_train)
decision_tree_model = DecisionTreeClassifier().fit(X_train, y_train)
random_forest_model = RandomForestClassifier().fit(X_train, y_train)
```

```python
#  predictions on the test set
y_pred_logistic_regression = logistic_regression_model.predict(X_test)
y_pred_knn = knn_model.predict(X_test)
y_pred_svm = svm_model.predict(X_test)
y_pred_naive_bayes = naive_bayes_model.predict(X_test)
y_pred_decision_tree = decision_tree_model.predict(X_test)
y_pred_random_forest = random_forest_model.predict(X_test)
```

```python
#  accuracy for each model
accuracy_logistic_regression = accuracy_score(y_test, y_pred_logistic_regression)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
accuracy_naive_bayes = accuracy_score(y_test, y_pred_naive_bayes)
accuracy_decision_tree = accuracy_score(y_test, y_pred_decision_tree)
accuracy_random_forest = accuracy_score(y_test, y_pred_random_forest)
```

```python
# Rounding off the accuracy scores to 2 decimal places
accuracy_logistic_regression = round(accuracy_logistic_regression, 2)
accuracy_knn = round(accuracy_knn, 2)
accuracy_svm = round(accuracy_svm, 2)
accuracy_naive_bayes = round(accuracy_naive_bayes, 2)
accuracy_decision_tree = round(accuracy_decision_tree, 2)
accuracy_random_forest = round(accuracy_random_forest, 2)
```

```python
# Print the accuracy scores
print("Accuracy Scores:")
print("Logistic Regression:", accuracy_logistic_regression)
print("K-nearest neighbor (KNN):", accuracy_knn)
print("Support Vector Machines (SVM):", accuracy_svm)
print("Naive Bayes:", accuracy_naive_bayes)
print("Decision Tree Classifier:", accuracy_decision_tree)
print("Random Forest Classifier:", accuracy_random_forest)
```

```python
# Create a dictionary to store the accuracy scores
accuracy_scores = {
    'Logistic Regression': accuracy_logistic_regression,
    'KNN': accuracy_knn,
    'SVM': accuracy_svm,
    'Naive Bayes': accuracy_naive_bayes,
    'Decision Tree': accuracy_decision_tree,
    'Random Forest': accuracy_random_forest,
}
```

```python
# Creating a bar plot
plt.figure(figsize=(11, 7))
plt.bar(accuracy_scores.keys(), accuracy_scores.values(), color='cyan')

# Adding labels and title
plt.xlabel('Machine Learning Models')
plt.ylabel('Accuracy Score')
plt.title('Comparison of Accuracy Scores for Different Models')

# Show the plot
plt.show()
```
